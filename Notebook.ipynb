# **Preprocessing**

```python
from sklearn.pipeline import Pipeline
import utils
from importlib import reload
reload(utils)
#We remove identifiers
df=data.drop(columns='Id') 

numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
categoric_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Fixing some mistakes such as MSSubClass, which should be in categoric_cols due to it's information. 
numeric_cols.remove('MSSubClass')
categoric_cols.append('MSSubClass')

display(numeric_cols)

```

## Treating date columns

```python

from utils import treat_date_cols
#Date columns in numeric must be treated as Dates, as it will benefit the future model
print(type(df['MoSold'][0]))
date_cols=['YearBuilt','YearRemodAdd','GarageYrBlt','MoSold','YrSold']

display(df[date_cols])
df_2=df

df_dated,date_cols,numeric_cols,categoric_cols=treat_date_cols(df_2,date_cols,numeric_cols,categoric_cols)
```

## Categoric Columns 

```python
df_dated['MSSubClass']=df['MSSubClass'].astype('category')
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for col in categoric_cols: 
    df_dated[col]=encoder.fit_transform(df_dated[col])
display(df_dated)

```

# EDA

```python
import matplotlib.pyplot as plt  
import seaborn as sns 
cor=df_dated.corr() 
# cor_abs = cor.abs()

# mostramos gráficamente
plt.figure(figsize=(12,10))
sns.heatmap(cor, annot=False, cmap=plt.cm.bwr)
plt.show()

# Filtramos las correlaciones respecto a 'SalePrice'
cor_saleprice = cor[['SalePrice']].sort_values(by='SalePrice', ascending=False)

# Creamos un heatmap
plt.figure(figsize=(8, 12))
sns.heatmap(cor_saleprice, annot=True, cmap=plt.cm.bwr, cbar=True)
plt.title("Correlaciones con SalePrice")
plt.show()
```
